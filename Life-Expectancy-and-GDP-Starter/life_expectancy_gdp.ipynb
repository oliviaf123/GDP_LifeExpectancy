```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df = pd.read_csv("all_data.csv")
print(df.head())
print(df.info())
print(df.Country.unique())
print(df.Year.unique())
print(df.columns)
print(df[df.isnull().any(axis=1)])




```

# Data Information
Within the provided data frame from the World Health Organization and World Bank there are 4 columns and 96 rows of data. The country column has 6 unique values: Chile, China, Germany, Mexico, United States of America, and Zimbabwe. The Year column has values ranging from 2000 to 2015 (inclusive) in 1 year increments. The Life Expectancy at Birth column is measured in years, and the GDP column stands for Gross Domestic Product. The Gross Domestic Product is a standard measure of the value added through the production of goods and services over a designated period of time, applied here, per year. It should also be noted that there are no null values.

```python
#the largest and most inclusive graph
sns.scatterplot(data=df, x=df['Life expectancy at birth (years)'], y=df['GDP'], hue=df['Country'])
plt.xlabel('Life expectancy at birth (in years)')
plt.ylabel('GDP')
plt.title('GDP vs. Life Expectancy')
plt.show()
plt.clf()
# print(df['Life expectancy at birth (years)'][df.Country == 'Germany'].max())
# print(df['Life expectancy at birth (years)'][df.Country == 'Chile'].max())
# print(df['Life expectancy at birth (years)'][df.Country == 'United States of America'].max())
```

# Bivariate Analysis
As seen in the graph above zimbabwe has a low GDP and a short life expectancy. The United States has the highest overall GDP. Germany, Chile, and the United States of America have the greatest life expectancy.

```python
#preparing the data
years = df.Year.unique()
ChileLE = df['Life expectancy at birth (years)'][df.Country == 'Chile']
ChinaLE = df['Life expectancy at birth (years)'][df.Country == 'China']
GermanyLE = df['Life expectancy at birth (years)'][df.Country == 'Germany']
MexicoLE = df['Life expectancy at birth (years)'][df.Country == 'Mexico']
USALE = df['Life expectancy at birth (years)'][df.Country == 'United States of America']
ZimbabweLE = df['Life expectancy at birth (years)'][df.Country == 'Zimbabwe']

#making a Life expectency at birth over the years graph
plt.plot(years,ChileLE, label='Chile')
plt.plot(years, ChinaLE, label='China')
plt.plot(years,GermanyLE, label='Germany')
plt.plot(years,MexicoLE,label='Mexico')
plt.plot(years,USALE,label='USA')
plt.plot(years,ZimbabweLE,label='Zimbabwe')
plt.xlabel("Year")
plt.ylabel("Life Expectancy at Birth in years")
plt.title('Life Expectancy at Birth over time per country')
plt.legend()
plt.show()
plt.clf()
```

# Univariate Analysis Part 1
The graph above shows that Zimbabwe has had the most significant increase in life expectancy at birth compared to the other countries. Germany has always had the highest life expectancy at birth for the time it was tracked in the provided data. Chile, China, Germany, Mexico, and the United States of America have life expectancies at birth that fall within 10 years of each other. Zimbabwe has a signifcantly lower life expectancy at birth as compared to the other 5 countries included.

```python
#preparing the data (AGAIN)
ChileGDP = df['GDP'][df.Country == 'Chile']
ChinaGDP = df['GDP'][df.Country == 'China']
GermanyGDP = df['GDP'][df.Country == 'Germany']
MexicoGDP = df['GDP'][df.Country == 'Mexico']
USAGDP = df['GDP'][df.Country == 'United States of America']
ZimbabweGDP = df['GDP'][df.Country == 'Zimbabwe']
#making the graph
plt.plot(years,ChileGDP,label='Chile')
plt.plot(years,ChinaGDP,label='China')
plt.plot(years, GermanyGDP, label='Germany')
plt.plot(years,MexicoGDP,label='Mexico')
plt.plot(years,USAGDP,label='USA')
plt.plot(years,ZimbabweGDP,label='Zimbabwe')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP per country from 2000 to 2015')
plt.legend()
plt.show()
plt.clf()
```

# Univariate Analysis part 2
The graph above shows the gross domestic product of each country from 2000 to 2015. As seen in the graph the United States of America's GDP has the highest GDP but it did dip around 2008/2009 presumably because of the large economic downturn at the time. Chile and Zimbabwe have the lowest and flattest GDPs over the time period. China's GDP has shown a signficant increase from 2000 to 2015.
